# Lekcja 5: Paginacja i filtrowanie API


## Cele
- params
- paginacja
- łączenie stron


```python
import requests
r=requests.get("https://reqres.in/api/users",params={"page":1},timeout=10)
p=r.json()
print(p["page"], p["total_pages"], len(p["data"]))
```

```python
def pobierz_wszystkich_uzytkownikow()->list[dict]:
    users=[]; page=1
    while True:
        d=requests.get("https://reqres.in/api/users",params={"page":page},timeout=10).json()
        users.extend(d["data"])
        if page>=d["total_pages"]: break
        page+=1
    return users
wszyscy=pobierz_wszystkich_uzytkownikow(); print(len(wszyscy))
```

```python
def filtruj_po_domenie(users:list[dict],domena:str)->list[dict]:
    return [u for u in users if u["email"].endswith(domena)]
print(len(filtruj_po_domenie(wszyscy,"reqres.in")))
```
