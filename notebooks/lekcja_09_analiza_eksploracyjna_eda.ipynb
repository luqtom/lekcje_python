## Cele
- statystyki
- grupowanie
- korelacje


```python
import pandas as pd
df=pd.DataFrame({"kategoria":["A","A","B","B","C"],"wartosc":[10,14,7,9,30],"czas_realizacji":[2,3,5,4,8]})
print(df.head())
```

```python
print(df.describe())
print(df.groupby("kategoria")["wartosc"].mean())
```

```python
print(df[["wartosc","czas_realizacji"]].corr())
```
