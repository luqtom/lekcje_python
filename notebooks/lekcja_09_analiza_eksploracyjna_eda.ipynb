## Zakres

- wykrywanie outlierów


```python
import pandas as pd

df = pd.DataFrame({"kategoria": ["A", "A", "B"], "wartosc": [10, 15, 7]})
print(df.groupby("kategoria")["wartosc"].mean())
```
