# Lekcja 4: Wprowadzenie do HTTP i biblioteki requests

Cel: pobierać dane z prostych endpointów API.


## Zakres
- metody HTTP (GET, POST)
- kody statusu
- nagłówki i timeout


```python
import requests

url = "https://jsonplaceholder.typicode.com/posts/1"
odp = requests.get(url, timeout=10)
print(odp.status_code)
print(odp.json()["title"])
```
