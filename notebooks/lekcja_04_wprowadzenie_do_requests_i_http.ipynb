# Lekcja 4: HTTP i `requests`


## Cele
- GET
- statusy i nagłówki
- JSON


```python
import requests
resp=requests.get("https://jsonplaceholder.typicode.com/posts/1",timeout=10)
print(resp.status_code, resp.headers.get("Content-Type"))
print(resp.json().keys())
```

```python
resp=requests.get("https://jsonplaceholder.typicode.com/posts",params={"userId":1},timeout=10)
posts=resp.json()
print(len(posts), posts[0]["title"])
```

```python
def pobierz_json(url:str):
    r=requests.get(url,timeout=10)
    r.raise_for_status()
    return r.json()
print(pobierz_json("https://jsonplaceholder.typicode.com/users/1")["email"])
```
