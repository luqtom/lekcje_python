## Cele
- zapis i odczyt JSON
- CSV DictWriter/DictReader


```python
import json
dane={"source":"api","records":3,"ok":True}
with open("sample.json","w",encoding="utf-8") as f: json.dump(dane,f,ensure_ascii=False,indent=2)
with open("sample.json","r",encoding="utf-8") as f: print(json.load(f))
```

```python
import csv
rows=[{"id":1,"city":"Warszawa","sales":120},{"id":2,"city":"Kraków","sales":95}]
with open("sales.csv","w",newline="",encoding="utf-8") as f:
    w=csv.DictWriter(f,fieldnames=["id","city","sales"])
    w.writeheader(); w.writerows(rows)
with open("sales.csv","r",encoding="utf-8") as f:
    for r in csv.DictReader(f): print(r)
```
