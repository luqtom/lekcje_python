```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.DataFrame({"miesiac":["sty","lut","mar","kwi"],"sprzedaz":[100,115,108,130],"koszt":[70,72,74,80]})
plt.plot(df["miesiac"],df["sprzedaz"],marker="o",label="Sprzedaż")
plt.plot(df["miesiac"],df["koszt"],marker="o",label="Koszt")
plt.legend(); plt.title("Trend miesięczny"); plt.show()
```

```python
sns.barplot(data=df,x="miesiac",y="sprzedaz"); plt.title("Sprzedaż wg miesiąca"); plt.show()
```
