````python
import logging
logging.basicConfig(level=logging.INFO,format="%(asctime)s | %(levelname)s | %(message)s")
logging.info("Start pipeline")
def policz_vat(cena_netto:float,stawka:float=0.23)->float:
    return round(cena_netto*(1+stawka),2)
logging.info("Cena brutto: %s",policz_vat(100))
````

````python
assert policz_vat(100)==123.0
assert policz_vat(0)==0.0
print("Testy lokalne OK")
````
