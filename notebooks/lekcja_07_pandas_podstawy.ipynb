```python
import pandas as pd
df=pd.DataFrame({"miasto":["Warszawa","Gdańsk","Kraków","Poznań"],"sprzedaz":[120,95,110,99],"miesiac":["styczeń"]*4})
print(df)
```

```python
print(df[["miasto","sprzedaz"]])
print(df[df["sprzedaz"]>100])
```

```python
print(df["sprzedaz"].describe())
print(df["sprzedaz"].sum())
```
