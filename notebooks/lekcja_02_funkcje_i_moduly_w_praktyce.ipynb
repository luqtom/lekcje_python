```python
def normalizuj_tekst(tekst: str) -> str:
    # usuń nadmiarowe spacje i zamień na małe litery
    return " ".join(tekst.strip().lower().split())

print(normalizuj_tekst("  DANE   API  "))
```

```python
def policz_srednia(liczby: list[float]) -> float:
    if not liczby:
        return 0.0
    return sum(liczby)/len(liczby)
print(policz_srednia([1.5,2.5,3.0]))
```

```python
def parse_price(value:str)->float:
    return float(value.replace(",", ".").replace("zł", "").strip())
print(parse_price("19,99 zł"))
```
