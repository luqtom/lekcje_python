```python
import os
API_TOKEN=os.getenv("API_TOKEN")
print("Token ustawiony:", bool(API_TOKEN))
```

```python
import requests
def pobierz_z_autoryzacja(url:str,token:str|None):
    headers={"Authorization":f"Bearer {token}"} if token else {}
    r=requests.get(url,headers=headers,timeout=10)
    print(r.status_code)
    return r
_ = pobierz_z_autoryzacja("https://jsonplaceholder.typicode.com/todos/1",API_TOKEN)
```
