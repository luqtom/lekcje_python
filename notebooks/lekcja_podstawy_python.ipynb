## 2. Operacje na liczbach

```python
a = 10
b = 3

print("Dodawanie:", a + b)
print("Dzielenie całkowite:", a // b)
print("Potęgowanie:", a ** b)
```

## 3. Instrukcje warunkowe

```python
ocena = 4

if ocena >= 5:
    print("Bardzo dobrze!")
elif ocena >= 3:
    print("Zaliczone.")
else:
    print("Do poprawy.")
```

## 4. Pętle

```python
for i in range(1, 6):
    print(f"Iteracja {i}")
```

## 5. Funkcje

```python
def pole_prostokata(a, b):
    return a * b

print("Pole prostokąta 3x7:", pole_prostokata(3, 7))
```
