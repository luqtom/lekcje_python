# Lekcja 8: Czyszczenie i transformacja danych


## Cele
- fillna
- konwersja typów
- kolumny pochodne


```python
import pandas as pd
df=pd.DataFrame({"produkt":["A","B","C",None],"cena":["10.0",None,"30.5","25.0"],"sztuki":[2,1,None,3]})
print(df)
```

```python
df["produkt"]=df["produkt"].fillna("NIEZNANY")
df["cena"]=pd.to_numeric(df["cena"],errors="coerce").fillna(0)
df["sztuki"]=df["sztuki"].fillna(0).astype(int)
print(df)
```

```python
df["wartosc"]=df["cena"]*df["sztuki"]
print(df[["produkt","wartosc"]])
```
